```python
#LR is the regression model computes the linear relationship between inputs and output
#LR has 3 types simple LR, multi Lr and polinomial Lr(mainly used when data is not in linear or sort of linear relatonship)
#Main goal is to find the best fit line for the data which tends to give minimal error between actual and predicted values using parameter slope and intercept
# line-> Y = M*X + B 
```

```python
import pandas as pd
```

```python
df=pd.read_csv("//Users//udayladdha//Desktop//Data science//DataSets//placement.csv")
df
```

```python
# package is our target colum 
# as this is a regression problem target column has continous numercial data  
```

```python
import matplotlib.pyplot as plt
```

```python
# as we can see these two have sort of linear regression so we can apply linear regression over here
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
lr.fit(x_train,y_train)
```

```python
x_test[0:2]
```

```python
y_test[0:2]
```

```python
# now i will give my model cgpa of 112 and 29 , then i will check how close they are having their pakage
```

```python
lr.predict(x_test.iloc[[0]])
# according to the algo the guy with cgpa 8.58 should have package of 3.89 instead of 4.10
```

```python
lr.predict(x_test.iloc[[1]])#according to the algo the guy with cgpa 7.15 should have package of 3.09 instead of 3.49
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.plot(x_train,lr.predict(x_train),color="red")
plt.xlabel="Cgpa"
plt.ylabel="Package"
plt.show()
```

```python
# LR using slopes and intercept to find the best fit line 
```
