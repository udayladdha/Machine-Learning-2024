```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = 6 * np.random.rand(100, 1) - 3
y =0.5 * X**2 + 1.5*X + 2 + np. random.rand(100, 1)
```

```python
plt.scatter(X,y,color="green")
plt.xlabel("X_feature")
plt.ylabel("Y_feature")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test=train_test_split(X,y,test_size=0.2, random_state=32)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,lr.predict(x_test))
```

```python
score
```

```python
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,lr.predict(x_train),"b")
plt.xlabel("X data")
plt.ylabel("Y data")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
```

```python
poly_x_train=poly.fit_transform(x_train)
poly_x_test=poly.transform(x_test)
```

```python
regression=LinearRegression()
regression.fit(poly_x_train,y_train)
y_pred=regression.predict(poly_x_test)
r2_score(y_test,y_pred)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.scatter(x_train,regression.predict(poly_x_train))
plt.scatter(x_train,y_train)
```

```python
x1=4 * np.random.rand(100,1)  - 2
x2=4 * np.random.rand(100, 1) - 2
x3=4 * np.random.rand(100, 1) - 2
```

```python
X = np.hstack((x1, x2, x3))
```

```python
Y= 0.8* X**2 + 0.9*X + 2 + np. random. rand(100, 1)
```

```python
plt.scatter(X,Y,color="purple")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2,include_bias=True)
```

```python
x_train , x_test , y_train , y_test=train_test_split(X,Y,test_size=0.2, random_state=32)
```

```python
x_train_transformed=pf.fit_transform(x_train)
x_test_transformed=pf.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_transformed,y_train)
y_pred=lr.predict(x_test_transformed)
r2_score(y_test,y_pred)
```
